```python
data=pd.read_csv("F:/kaggel/HAM10000_metadata.csv")
data.head()
```

```python
data.info()
```

```python
 data.drop(['lesion_id', 'image_id'], axis=1,inplace=True)
data.info()
```

```python
data.age.value_counts()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(data.age)
plt.title("Age of cancer patient",size=20)
plt.show()  
```

```python
plt.figure(figsize=(8,8))
sns.countplot(data.sex)
plt.title("gender of cancer patient",size=15)
plt.show()
```

```python
plt.figure(figsize=(20,15))
sns.countplot(x='sex',hue='age',data=data)
plt.legend(loc='best')
plt.title("Age & gender of cancer patient",size=15)
plt.show()
```

```python
labels=data.localization.value_counts().index
sizes=data.localization.value_counts().values
plt.figure(figsize=(11,11))
plt.pie(sizes,labels=labels,autopct="%1.f%%")
plt.title("cancer location",size=25)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x=labels,y=sizes)
plt.show()
```

```python
labels1=data.dx.value_counts().index
sizes1=data.dx.value_counts().values
plt.figure(figsize=(11,11))
plt.pie(sizes1,labels=labels1,autopct="%1.1f%%")
plt.title("seven most common types of skin cancer",size=25)
plt.show()
```
